```python
import pandas as pd
import numpy as np

df_bikes = pd.read_csv('../input/data-wrangling-bike-rentals/bike_rentals_cleaned.csv')

df_bikes.head()
```

```python
X = df_bikes.iloc[:,:-1]
y = df_bikes.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 2)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error

dec_reg = DecisionTreeRegressor(random_state=2)
params = {'max_depth':[None, 2,3,4,6,8,10,20]}

grid_reg = GridSearchCV(dec_reg, params, scoring='neg_mean_squared_error', cv=5, n_jobs=-1)

grid_reg.fit(X_train,y_train)
```

```python
best_params = grid_reg.best_params_
print('Best Params:',best_params)
```

```python
best_score = np.sqrt(-grid_reg.best_score_)
print('Training Score: %0.3f'% best_score)
```

```python
best_model = grid_reg.best_estimator_
y_preds = best_model.predict(X_test)

rmse_test = mean_squared_error(y_preds, y_test)**0.5
print('RMSE: ',np.round(rmse_test,3))
```

```python
def grid_search(params, reg=DecisionTreeRegressor(random_state=2)):
    grid_reg = GridSearchCV(reg, params,scoring='neg_mean_squared_error', cv=5, n_jobs=-1)
    grid_reg.fit(X_train,y_train)
    best_params = grid_reg.best_params_
    print('Best Params: ',best_params)
    best_score = np.sqrt(-grid_reg.best_score_) 
    print("Training score:{:.3f}".format(best_score))
    y_preds = grid_reg.predict(X_test)
    rmse_test = mean_squared_error(y_preds, y_test)**0.5
    print('RMSE: {:.3f}'.format(rmse_test))
```

```python
grid_search(params={'min_samples_leaf':[1,2,4,6,8,10,20,30]})
```

```python
grid_search(params={'max_depth':[None,2,3,4,6,8,10,20],
                   'min_samples_leaf':[1,2,4,6,8,10,20,30]})
```

```python
grid_search(params={'max_depth':[6,7,8,9,10],
                   'min_samples_leaf':[3,5,7,9]})
```
